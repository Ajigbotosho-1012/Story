```python
# code snipper to be executed only once

setup = '''
import numpy as np
import pandas as pd
df = pd.read_csv('tips.csv')
def quality(total_bill, tip):
    if tip/total_bill > 0.25:
        return "Generous"
    else:
        return "Other"
'''
```

```python
# code snippet whose execution time is to be measured


stmt_two = '''
df['Tip Quality'] = np.vectorize(quality)(df['total_bill'],df['tip'])
'''
```

```python
timeit.timeit(setup=setup,stmt=stmt_two,number=1000)
```

```python
stmt_one = """
df['Tip Quality'] = df[['total_bill','tip']].apply(lambda df:
                              quality(df['total_bill'],
                                     df['tip']),
                              axis = 1)
"""
```

```python
timeit.timeit(setup=setup,stmt=stmt_one,number=1000)
```
